```python
city_Chicago = gpd.read_file('https://data.cityofchicago.org/resource/p293-wvbd.geojson')
city_Chicago
```

```python
homocides_index["ward_id"]=homocides_index.index
```

```python
homocides_index.info()
```

```python

city_Chicago["ward_id"]= city_Chicago["ward_id"].astype("float")
city_Chicago.info()
```

```python
outer_merge_df = pd.merge(homocides_index,city_Chicago,on= "ward_id",how="right")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt


#This map was not used on site 
# GeoDataFrame called city_Chicago with a 'geometry' column containing ward boundaries
# Outer_merge_df contains the data you want to merge with the ward boundaries

# Convert the 'geometry' column in city_Chicago to a list of polygons
polygons = list(city_Chicago['geometry'])

# Create a GeoDataFrame from outer_merge_df with the geometry column populated by the list of polygons
merged_gdf = gpd.GeoDataFrame(outer_merge_df, geometry=polygons)

# Plot the GeoDataFrame
fig, ax = plt.subplots(figsize=(7, 7))
merged_gdf.plot(column='homocide_count', cmap='coolwarm', legend=True, ax=ax)

# Add text labels for ward numbers at centroids
for idx, row in merged_gdf.iterrows():
    centroid = row['geometry'].centroid
    ward_number = row['ward_id']
    ax.text(centroid.x, centroid.y, str(ward_number), fontsize=8, ha='center', va='center')

plt.title('Chicago Homicide Counts by Ward (2002-Present)')
plt.show()
```

From the cell below onward I was able to concetenate my merge into the single cell block. All of the year over year maps represent homocide count by ward number with the  darkest red being the most dangerous (highest count ) to the darkest blue being the safest (lowest count)
