```python
# Import dependencies
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
#Create a dataframe of the crime data
crimeData = pd.read_csv(Path("Resources/crime_data.csv"))
crimeData.head()
```

```python
#Limit data to 2017-2023
crimeData = crimeData[(crimeData["Year"] >= 2017) & (crimeData["Year"] <= 2023)]
```

```python
topCrimes = crimeData.value_counts('Primary Type')[0:10].plot(kind='bar')
topCrimes.set_title("Top 10 Most Frequent Crimes")
topCrimes.set_xlabel("Primary Type of Crime")
topCrimes.set_ylabel("Frequency")
```

## Crime Rates Between Seasons ##

```python
# Create seperate dataframes for each season
winterCrimeData = crimeData[(crimeData['Date'].str.split('/').str[0] == "12") | (crimeData['Date'].str.split('/').str[0] == "01") | (crimeData['Date'].str.split('/').str[0] == "02")]
springCrimeData = crimeData[(crimeData['Date'].str.split('/').str[0] == "03") | (crimeData['Date'].str.split('/').str[0] == "04") | (crimeData['Date'].str.split('/').str[0] == "05")]
summerCrimeData = crimeData[(crimeData['Date'].str.split('/').str[0] == "06") | (crimeData['Date'].str.split('/').str[0] == "07") | (crimeData['Date'].str.split('/').str[0] == "08")]
fallCrimeData = crimeData[(crimeData['Date'].str.split('/').str[0] == "09") | (crimeData['Date'].str.split('/').str[0] == "10") | (crimeData['Date'].str.split('/').str[0] == "11")]
```

```python
# Top 10 crimes in winter
topWinter = winterCrimeData.value_counts('Primary Type')[0:10].plot(kind='bar')
topWinter.set_title("Top 10 Crimes in Winter")
topWinter.set_xlabel("Primary Type of Crime")
topWinter.set_ylabel("Frequency")
topWinter
```

```python
# Top 10 crimes in spring
topSpring = springCrimeData.value_counts('Primary Type')[0:10].plot(kind='bar')
topSpring.set_title("Top 10 Crimes in Spring")
topSpring.set_xlabel("Primary Type of Crime")
topSpring.set_ylabel("Frequency")
topSpring
```

```python
# Top 10 crimes in summer
topSummer = summerCrimeData.value_counts('Primary Type')[0:10].plot(kind='bar')
topSummer.set_title("Top 10 Crimes in Summer")
topSummer.set_xlabel("Primary Type of Crime")
topSummer.set_ylabel("Frequency")
topSummer
```

```python
# Top 10 crimes in fall
topFall = fallCrimeData.value_counts('Primary Type')[0:10].plot(kind='bar')
topFall.set_title("Top 10 Crimes in Fall")
topFall.set_xlabel("Primary Type of Crime")
topFall.set_ylabel("Frequency")
topFall
```

```python
# Number of crimes per year in winter
winterYears = winterCrimeData.value_counts('Year').sort_index().plot(kind='bar')
winterYears.set_title("Number of Crimes Per Year in Winter")
winterYears.set_xlabel("Year")
winterYears.set_ylabel("Frequency")
winterYears
```

```python
# Number of crimes per year in spring
springYears = springCrimeData.value_counts('Year').sort_index().plot(kind='bar')
springYears.set_title("Number of Crimes Per Year in Spring")
springYears.set_xlabel("Year")
springYears.set_ylabel("Frequency")
springYears
```

```python
# Number of crimes per year in summer
summerYears = summerCrimeData.value_counts('Year').sort_index().plot(kind='bar')
summerYears.set_title("Number of Crimes Per Year in Summer")
summerYears.set_xlabel("Year")
summerYears.set_ylabel("Frequency")
summerYears
```

```python
# Number of crimes per year in fall
fallYears = fallCrimeData.value_counts('Year').sort_index().plot(kind='bar')
fallYears.set_title("Number of Crimes Per Year in Fall")
fallYears.set_xlabel("Year")
fallYears.set_ylabel("Frequency")
fallYears
```

```python
# Number of crimes per season
plt.bar(["Winter","Spring","Summer","Fall"], [len(winterCrimeData.index), len(springCrimeData.index), len(summerCrimeData.index), len(fallCrimeData.index)])
plt.title("Number of Crimes Per Season")
plt.xlabel("Season")
plt.ylabel("Frequency")
plt.show()
```

## Crime Time of Day Distributions ##

```python
times = []
for time in crimeData['Date'].str.split(' ').str[1:3].tolist():
    if time[1] == 'PM' and time[0][0:2] != '12':
        times.append(int(time[0][0:2]) + int(time[0][3:5])/60 + int(time[0][6:8])/360 + 12)
    else:
        times.append(int(time[0][0:2]) + int(time[0][3:5])/60 + int(time[0][6:8])/360)
times
```

```python
Counter(times)
```

```python
#Time distribustion of crimes
plt.plot(*zip(*sorted(Counter(times).items())))
plt.xticks([i for i in range(0,25)])
plt.subplots_adjust(right=1.5)
plt.title("Time Distribution of Crimes")
plt.xlabel("Hours after midnight")
plt.ylabel("Frequency")
plt.show
```
