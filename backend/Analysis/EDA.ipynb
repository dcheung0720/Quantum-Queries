```python
import pandas as pd
import re
from datetime import datetime 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("../Data/Crimes2001_to_Present.csv")
df.head()
```

```python
df.info()
```

```python
#ignore the last year
crimes_per_year = df.groupby("Year").size().reset_index(name='Count').iloc[0:-1]
crimes_per_year.plot(x = "Year", y = "Count")
plt.title("Crime Rate vs. Year")
plt.xlabel("Year")
plt.ylabel("Crime Count")
plt.show()
```

```python
# All the different times of crimes
df["Primary Type"].unique()
```

```python
violent_crimes = ['BATTERY', 'ASSAULT', 'CRIM SEXUAL ASSAULT', 'ROBBERY', 'CRIMINAL SEXUAL ASSAULT', 'HOMICIDE', 'DOMESTIC VIOLENCE']
df_violent = df[df["Primary Type"].isin(violent_crimes)]

violent_crime_types = df_violent.groupby(["Primary Type", "Year"]).size().reset_index(name = "Count")
violent_crime_types = violent_crime_types[violent_crime_types["Year"] != 2024]
plt.figure(figsize=(10, 6))
sns.lineplot(data= violent_crime_types, x="Year", y="Count", hue="Primary Type", palette="tab10")
plt.legend(title='Primary Type')
plt.show()
```

```python
years = [2001 + i for i in range(23)]
data = {}
for crime in violent_crimes:
    data_crimes = []
    for year in years:
        data_crime = violent_crime_types[(violent_crime_types["Year"] == year) & (violent_crime_types["Primary Type"] == crime)]["Count"].values
        data_crime = data_crime[0] if len(data_crime) > 0 else 0
        data_crimes.append(data_crime)
    data[crime] = data_crimes

fig, ax = plt.subplots()
bottom = np.zeros(23)
for crime, counts in data.items():
    p = ax.bar(years, counts, 0.5, label=crime, bottom=bottom)
    bottom += counts

ax.set_title("Crimes vs. Time")
ax.legend(loc = "upper right")

plt.show()
```
