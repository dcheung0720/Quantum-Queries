Data Cleaning 


```python
df = pd.read_csv("/Users/eweli/Desktop/activity file/project 3/Crimes_-_2001_to_Present_20240408.csv")
```

```python
# Delate unecessary columns in DataFrame
arrest_df = df.drop(['Case Number','IUCR','Domestic','Block','IUCR','Primary Type','Description','Location Description','Domestic','Beat','District','Ward','Community Area','FBI Code','X Coordinate','Y Coordinate','Year','Updated On','Latitude','Longitude','Location'],axis = 1)
```

```python
# Drop all rows with missing informations
arrest_df = arrest_df.dropna(how='any')
```

```python
# Reorganize Columns
arrest_df = arrest_df[['Arrest','Date']]
```

```python
# Convert 'Date' column to datetime
arrest_df['Date'] = pd.to_datetime(arrest_df['Date'])

# Extracting time from 'Date' columns
arrest_df['Time'] = arrest_df['Date'].dt.time

# Extract month from 'Date' column and month number to abbreviated month name
arrest_df['Month'] = arrest_df['Date'].dt.month

arrest_df['Month'] = arrest_df['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
# Creat 'Year' column by extracting yaer from "Date" column
if 'Date' in arrest_df.columns:
    arrest_df['Year'] = arrest_df['Date'].dt.year
  
```

```python
# Reorganize Columns
arrest_df = arrest_df[['Arrest','Date','Year','Time']]
# Filter the data from 2017 to 2024
arrest_df = arrest_df[(arrest_df['Year'] >= 2017) & (arrest_df['Year'] <=2024)]
arrest_df
```

Arrests made over time

```python
arrest_df ['Arrest'].value_counts()
```

```python
# The colors of each section of the pie chart
colors = ["dimgrey","blue"]
labels = ['No Arrests', 'Arrests']

# Separate arrest made from no arrest 
explode = (0.1, 0)

# Set figure size
plt.figure(figsize=(10,6))

# Create the pie chart based on the arrests made
arrest_counts = arrest_df['Arrest'].value_counts()
total_arrests = arrest_counts.sum()
plt.pie(arrest_counts, labels=labels, colors=colors, explode=explode, autopct=lambda p: f'{int(p * total_arrests / 100)}', shadow=True, startangle=180)
plt.title('Arrests vs No Arrests')

# Annotate the pie chart with additional information
plt.text(-0.6, 0, f"Total: {total_arrests}\nArrests: {arrest_counts[True]}\nNo Arrests: {arrest_counts[False]}", fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

# Add the count values to the legend
plt.legend(labels=[f'{label} ({count})' for label, count in zip(labels, arrest_counts)], loc="upper left", fontsize=10)

plt.show()
```
